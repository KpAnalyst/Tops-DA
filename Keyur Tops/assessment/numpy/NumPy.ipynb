```python
"""6) What are the top 5 states with the most failed banks? """
s3 = banks["ST"].value_counts().nlargest(5)
print("The top 5 States with most failed banks is:",s3)
```

```python
""" 7) What are the top 5 acquiring institutions? """
s4 = banks["Acquiring Institution"].value_counts().nlargest(5)
print(s4)
```

```python
""" 8) How many banks has the State Bank of Texas acquired? How many of them were actually in Texas?"""
s5 = banks[banks["Acquiring Institution"] == "State Bank of Texas"].shape[0]
s6 = banks[(banks["Acquiring Institution"] == "State Bank of Texas") & (banks["ST"] == "TX")].shape[0]
print("State Bank of Texas acquired ",s5," banks, out of which", s6," were in Texas.")
```

```python
""" #9) What is the most common city in California for a bank to fail in? """
s7 = banks.loc[banks["ST"] == "CA", "City"].value_counts().idxmax()
print(s7)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
f = pd.read_csv(r"C:\Users\KEYUR\Desktop\Keyur Tops\assessment\numpy\historical_automobile_sales.csv")  
f
```

```python
"""Q 1: Develop a Line chart using the functionality of pandas to show how automobile sales fluctuate from year to year. """
p1 = f.groupby("Year")["Automobile_Sales"].sum()

p1.plot(kind="line",color="r", marker="*", figsize=(10, 5), title="Automobile Sales Trends Over the Years")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.grid()
plt.show()
```

```python
"""Q 2: Plot different lines for categories of vehicle type and analyze the trend to answer the question Is there a noticeable difference in 
sales trends between different vehicle types during recession periods?"""
p2= f.groupby(["Year", "Vehicle_Type"])["Automobile_Sales"].sum().unstack()
plt.figure(figsize=(12, 6))
for i in p2:
    plt.plot(p2.index, p2[i], marker="^", label=i)
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.title("Automobile Sales by Vehicle Type")
plt.legend(title="Vehicle Type")
plt.grid()
plt.show()
```

```python
"""Q 3: Use the functionality of Seaborn Library to create a visualization to compare the sales trend per vehicle type for
a recession period with a non- recession period."""
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(data=f, x="Year", y="Automobile_Sales", hue="Vehicle_Type", style="Recession", markers=["^","o"], dashes=False)
plt.xlabel("Year")
plt.ylabel("Sales")
plt.title("Sales Trend per Vehicle Type: Recession vs. Non-Recession")
plt.legend(title="Vehicle Type & Recession Status")
plt.grid()
plt.show()
```
