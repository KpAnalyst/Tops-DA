```python
""" 4) How many States (ST) are represented in this data set?"""
states = banks["ST"].nunique()
print(f"Number of unique states is: {states}")
```

```python
""" 5) Get a list or array of all the states in the data set."""
s_list = banks["ST"].unique()
print("List of states:\n",s_list)
```

```python
"""6) What are the top 5 states with the most failed banks? """
top_5 = banks["ST"].value_counts().head(5)
print("Top 5 states with most failed banks:")
print(top_5)
```

```python
""" 7) What are the top 5 acquiring institutions? """
top_acquirers = banks["Acquiring Institution"].value_counts().head(5)
print("Top 5 acquiring institutions:\n" + top_acquirers.to_string())
```

```python
""" 8) How many banks has the State Bank of Texas acquired? How many of them were actually in Texas?"""
tx_acq = banks[banks["Acquiring Institution"] == "State Bank of Texas"]
num_acq = tx_acq.shape[0]
texas = tx_acq[tx_acq["ST"] == "TX"].shape[0]

print(f"State Bank of Texas acquired {num_acq} banks, out of which {texas} were in Texas.")
```

```python
""" #9) What is the most common city in California for a bank to fail in? """
most_common = banks[banks["ST"] == "CA"]["City"].mode()[0]

print(f"The most common city in California for bank failures is {most_common}.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = pd.read_csv(r"C:\Users\KEYUR\Desktop\Keyur Tops\assignment\python\numpy\historical_automobile_sales.csv")  
f
```

```python
"""Q 1: Develop a Line chart using the functionality of pandas to show how automobile sales fluctuate from year to year. """
sales = f.groupby("Year")["Automobile_Sales"].sum()

plt.figure(figsize=(10, 5))
plt.plot(sales, marker="o", linestyle="-", color="b")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.title("Automobile Sales Trends Over the Years")
plt.grid()
plt.show()
```

```python
"""Q 2: Plot different lines for categories of vehicle type and analyze the trend to answer the question Is there a noticeable difference in 
sales trends between different vehicle types during recession periods?"""
vehicle_sales = f.groupby(["Year", "Vehicle_Type"])["Automobile_Sales"].sum().unstack()

plt.figure(figsize=(12, 6))
vehicle_sales.plot(marker="o", linestyle="-", colormap="viridis", figsize=(12, 6))
plt.xlabel("Year")
plt.ylabel("Sales")
plt.title("Sales Trends by Vehicle Type Over the Years")
plt.legend(title="Vehicle Type")
plt.grid()
plt.show()
```

```python
"""Q 3: Use the functionality of Seaborn Library to create a visualization to compare the sales trend per vehicle type for
a recession period with a non- recession period."""
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(data=f, x="Year", y="Automobile_Sales", hue="Vehicle_Type", style="Recession", markers=True, dashes=False)
plt.xlabel("Year")
plt.ylabel("Sales")
plt.title("Sales Trend per Vehicle Type: Recession vs. Non-Recession")
plt.legend(title="Vehicle Type & Recession Status")
plt.grid()
plt.show()
```

```python
#Q 4: Now you want to compare the sales of different vehicle types during a recession and a non-recession period
rec_sales = f.groupby(["Vehicle_Type", "Recession"])["Automobile_Sales"].sum().unstack()

rec_sales.plot(kind="bar", figsize=(10, 5), colormap="coolwarm")
plt.xlabel("Vehicle Type")
plt.ylabel("Total Sales")
plt.title("Vehicle Sales: Recession vs. Non-Recession Periods")
plt.legend(["Non-Recession", "Recession"])
plt.xticks(rotation=0)
plt.grid()
plt.show()
```
