```python
# Calculate the total sales and average price per product.
df = pd.DataFrame({
    "Product": ["A", "A", "B", "B"],
    "Sales": [100, 200, 150, 250],
    "Price": [10, 20, 15, 25]
})
df.groupby("Product").agg({"Sales":"sum","Price":"mean"})
```

```python
# Create a pivot table showing average salary by Department and Role.
df = pd.DataFrame({
    "Department": ["HR", "Tech", "HR", "Tech"],
    "Role": ["Manager", "Engineer", "Manager", "Engineer"],
    "Salary": [70, 80, 75, 85]
})
re=pd.pivot_table(df,values="Salary",index="Role",columns="Department",aggfunc="mean")
re
```

```python
# Convert the wide-format DataFrame to long format.
# syntex: pd.melt(frame, id_vars=None, value_vars=None, var_name=None, value_name='value')
# frame â†’ The DataFrame you want to melt.
# id_vars â†’ The columns that should remain fixed (not melted).
# value_vars â†’ The columns that should be melted into rows.
# var_name â†’ Name for the new column that will store former column names.
# value_name â†’ Name for the new column that will store values.
df = pd.DataFrame({"ID": [1, 2], "Math": [90, 85], "Science": [88, 92]})
df
melted = df.melt(id_vars="ID", var_name="Subject", value_name="Score")
melted
```

```python
# onvert the Date column from string to datetime format.
df = pd.DataFrame({"Date": ["2023-01-01", "2023-02-01"]})
df
df["Date"]=pd.to_datetime(df["Date"])
df
```

```python
# Add a column for the month from the Date column.
df = pd.DataFrame({"Date": pd.date_range(start="2023-01-01", periods=3, freq="ME")})
df["Month"]=df["Date"].dt.month
df
```

```python
# Resample daily sales data to monthly totals.dates = pd.date_range(start="2023-01-01", periods=5, freq="D")
df = pd.DataFrame({"Date": dates, "Sales": [10, 20, 30, 40, 50]})
df
M_t=df.set_index("Date").resample("ME").sum()
M_t
```
